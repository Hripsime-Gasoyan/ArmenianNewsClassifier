```python
from selenium import webdriver
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import requests
import csv
import time
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
```

```python
# Create an instance of the WebDriver
driver = webdriver.Chrome("/chromedriver")

# # # Navigate to the website
url = 'https://www.azatutyun.am/z/2041' 
# Replace with the URL of the website you want to scrape
driver.get(url)

# # Wait for the "Load More" button to be clickable
show_more_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, '#content > div:nth-child(1) > div > div > div > div.col-xs-12.col-md-8.col-lg-8.pull-left.content-offset > div.media-block-wrap > p > a')))

# Wait for the additional content to load (you may need to adjust the wait time based on the page)
WebDriverWait(driver, 10).until(EC.invisibility_of_element_located((By.CSS_SELECTOR, '.loader')))

# Extract the updated content
updated_content = driver.find_element(By.CSS_SELECTOR, '#content').get_attribute('innerHTML')

# Click the "Load More" button repeatedly until all content is loaded
while True:
    try:
        # Click the "Load More" button
        show_more_button.click()
        time.sleep(2)  # Add a small delay to allow the content to load
    except:
        break  # Break the loop if the button is no longer clickable or there is no more content to load

# Parse the HTML content
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
div_elements = soup.find_all("div", class_="media-block__content media-block__content--h media-block__content--h-xs")
```

```python
titles = []

# Iterate over the div elements
for div_element in div_elements:
    # Find all the h4 elements within the div
    h4s = div_element.find_all('h4')

    # Iterate over the h4 elements
    for h4 in h4s:
        # Extract the title attribute
        title = h4.get('title')
        # Append the title to the list
        titles.append(title)
        
csv_file = 'տնտեսություն.csv'

# Write the titles to a CSV file
with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Title'])  # Write the header row
    
    for title in titles:
        writer.writerow([title])


# Close the browser
driver.quit()
```

```python
# Create a CSV file and open it in write mode
with open('տնտեսություն_տեքստ.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Text'])
    for div_element in div_elements:
        news = div_element.find_all("a")
        for n1 in news:
            url1 = 'https://www.azatutyun.am/' + n1.get("href")
            response1 = requests.get(url1)
            soup1 = BeautifulSoup(response1.content, 'html.parser')
            text = " "
            div_elements1 = soup1.find_all("div", class_="wsw") 
            for div_element1 in div_elements1:
                    p_elements = div_element1.find_all("p")
                    for p_element in p_elements:
                        # Append the text content to the list
                        text += p_element.text
                    # Write the text content into a row in the CSV file
            writer.writerow([text])
```
